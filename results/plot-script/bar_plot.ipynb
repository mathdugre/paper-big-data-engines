```python
import plot

col_names = ["func", "start", "end", "filename", "worker", "thread", "process"]
benchmark_dir = f"../../results/benchmarks"
ylim=None
```

# Increment

```python
experiment = "increment"
parameters = ["n_worker", "iterations", "delay"]
```

```python
kwargs = {"iterations": 10, "delay": 1.0,}
xlabel = "# of Worker"

plot.bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    **kwargs,
    title="Increment worker",
    ylim=ylim,
)
```

```python
kwargs = {"delay": 1.0, "n_worker": 4,}
xlabel = "# of Iterations"

plot.bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    **kwargs,
    title="Increment iteration",
    ylim=ylim,
)
```

```python
kwargs = {"iterations": 10, "n_worker": 4,}
xlabel = "Delay [s]"

plot.bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    **kwargs,
    title="Increment delay",
    ylim=ylim,
)
```
