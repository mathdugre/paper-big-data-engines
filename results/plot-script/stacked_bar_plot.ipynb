```python
import plot
from plot import idle_time, wasted_time

col_names = ["func", "start", "end", "filename", "worker", "thread", "process"]
benchmark_dir = f"../benchmarks"
ylim = None
```

```python
idle_func = idle_time
```

# Increment

```python
experiment = "increment"
parameters = ["n_worker", "iterations", "delay"]
```

```python
kwargs = {
    "iterations": 10,
    "delay": 1.0,
}
xlabel = "# of Worker"

plot.stacked_bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    idle_func=idle_func,
    **kwargs,
    title="Increment worker",
    ylim=ylim,
)
```

```python
kwargs = {
    "delay": 1.0,
    "n_worker": 4,
}
xlabel = "# of Iterations"

plot.stacked_bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    idle_func=idle_func,
    **kwargs,
    title="Increment iteration",
    ylim=ylim,
)
```

```python
kwargs = {
    "iterations": 10,
    "n_worker": 4,
}
xlabel = "Delay [s]"

plot.stacked_bar(
    benchmark_dir=benchmark_dir,
    experiment=experiment,
    col_names=col_names,
    parameters=parameters,
    xlabel=xlabel,
    idle_func=idle_func,
    **kwargs,
    title="Increment delay",
    ylim=ylim,
)
```
